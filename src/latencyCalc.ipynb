```python
import matplotlib.pyplot as plt
```

```python
# Probability values
p_values = [0, 0.2, 0.4, 0.8]
```

```python
# Latencies with caching
product_query_with = [4.22871, 4.94261, 5.08454, 6.36845]
order_buy_with = [0, 14.52089, 15.14781, 16.24354]
order_get_with = [0, 10.34376, 10.56752, 11.78032]
# Latencies without caching
product_query_without = [4.82267, 5.85645, 5.83243, 6.66454]
order_buy_without = [0, 15.883434, 15.922143, 16.67835]
order_get_without = [0, 10.76438, 11.66326, 11.95421]
```

```python
# Function to calculate percentage improvement due to caching
def calculate_improvement(with_cache, without_cache):
    return [(1 - (w / wo)) * 100 if wo != 0 else 0 for w, wo in zip(with_cache, without_cache)]
```

```python
# Calculate improvements
product_query_improvement = calculate_improvement(product_query_with, product_query_without)
order_buy_improvement = calculate_improvement(order_buy_with, order_buy_without)
order_get_improvement = calculate_improvement(order_get_with, order_get_without)
```

```python
# Create plots
plt.figure(figsize=(14, 7))

# Plotting latency with caching
plt.subplot(1, 2, 1)
plt.plot(p_values, product_query_with, marker='o', label='Product Query')
plt.plot(p_values, order_buy_with, marker='o', label='Order Buy')
plt.plot(p_values, order_get_with, marker='o', label='Order Get')
plt.title('With Caching')
plt.xlabel('Probability p')
plt.ylabel('Latency (ms)')
plt.legend()
plt.grid(True)
plt.ylim(0, 18)  # Set y-axis limits

# Plotting latency without caching
plt.subplot(1, 2, 2)
plt.plot(p_values, product_query_without, marker='o', label='Product Query')
plt.plot(p_values, order_buy_without, marker='o', label='Order Buy')
plt.plot(p_values, order_get_without, marker='o', label='Order Get')
plt.title('Without Caching')
plt.xlabel('Probability p')
plt.ylabel('Latency (sec)')
plt.legend()
plt.grid(True)
plt.ylim(0, 18)  # Set y-axis limits

plt.tight_layout()
plt.show()
```

```python
improvement_types = ["Product Query", "Order Buy", "Order Get"]
improvements = [product_query_improvement, order_buy_improvement, order_get_improvement]

for i, imp_type in enumerate(improvement_types):
    for p_val, imp in zip(p_values, improvements[i]):
        print(f"Probability: {p_val:.1f}, Request Type: {imp_type}, Improvement: {imp:.2f}%")
```
